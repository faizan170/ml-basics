```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
#from preprocessing import one_hot_encouder
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn import metrics
import matplotlib.pyplot as plt
#https://www.kaggle.com/sabasiddiqi/workbook-2-text-preprocessing-feature-extraction
```

```python
data = pd.read_csv('kdd_2.csv')
data.head()
```

```python
y_data_main = data['Provisore'].values
y_data = pd.get_dummies(data['Provisore']).values
print(y_data[0])
x_data = data.drop('Provisore',axis=1).values
```

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2

x_data = SelectKBest(chi2, k=23).fit_transform(x_data, y_data)
x_data.shape
```

```python
mms = MinMaxScaler(feature_range=(0,1))
x_data = mms.fit_transform(x_data)

print(x_data.shape)
```
